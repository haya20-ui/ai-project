## 1) Introduction: Instance-based learning and intuition

- **KNN** is a *lazy* learner: it stores the training data and defers generalization to query time.
- **Intuition:** similar inputs (neighbors) tend to have similar outputs.
- **Key design choices:** 
  - Number of neighbors (**k**)
  - **Distance metric** (e.g., Euclidean, Manhattan, Chebyshev)
  - **Feature scaling** (distances are sensitive to feature scale)
- **Tradeoffs:**
  - Smaller k → lower bias, higher variance; larger k → higher bias, lower variance.
  - Different metrics can change nearest-neighbor structure and decision boundaries.
  - Scaling is often necessary for fair distance comparisons.


## 2) Setup and Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris, load_wine
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

import warnings
warnings.filterwarnings('ignore')

RANDOM_STATE = 42
np.set_printoptions(suppress=True)
print("Libraries imported. RANDOM_STATE =", RANDOM_STATE)
```


## 3) Iris Dataset — Baseline KNN

```python
# load Iris and prepare data
iris = load_iris()
X = iris.data
y = iris.target
feature_names = iris.feature_names
target_names = iris.target_names
print("Iris shape:", X.shape, "| Classes:", list(target_names))

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=RANDOM_STATE, stratify=y)

# Scaling is important for KNN due to distance computations
scaler = StandardScaler()
X_train_s = scaler.fit_transform(X_train)
X_test_s  = scaler.transform(X_test)

# Baseline KNN (k=5, Euclidean distance)
knn = KNeighborsClassifier(n_neighbors=5, metric='euclidean')
knn.fit(X_train_s, y_train)
y_pred = knn.predict(X_test_s)

print("Baseline (k=5, euclidean) — Accuracy:", round(accuracy_score(y_test, y_pred), 4))
print("\nClassification report:")
print(classification_report(y_test, y_pred, target_names=target_names))

cm = confusion_matrix(y_test, y_pred)
print("Confusion matrix:\n", cm)

# Simple confusion-matrix plot
plt.figure(figsize=(4,4))
plt.imshow(cm, interpolation='nearest')
plt.title("Iris — Confusion Matrix (k=5, euclidean)")
plt.colorbar()
tick_marks = np.arange(len(target_names))
plt.xticks(tick_marks, target_names, rotation=45)
plt.yticks(tick_marks, target_names)
plt.xlabel("Predicted")
plt.ylabel("True")
plt.tight_layout()
plt.show()
```


## 4) Compare Distance Metrics (fixed k=5)

We will evaluate **Euclidean**, **Manhattan**, and **Chebyshev** metrics with `k=5` and compare accuracies.


```python
# metrics comparison at k=5
metrics = ['euclidean', 'manhattan', 'chebyshev']
accs = []

for m in metrics:
    model = KNeighborsClassifier(n_neighbors=5, metric=m)
    model.fit(X_train_s, y_train)
    y_hat = model.predict(X_test_s)
    accs.append(accuracy_score(y_test, y_hat))

metrics_df = pd.DataFrame({"metric": metrics, "accuracy": accs}).sort_values("accuracy", ascending=False)
print(metrics_df.to_string(index=False))

# Plot accuracy by metric
plt.figure(figsize=(5,3.5))
plt.bar(metrics_df["metric"], metrics_df["accuracy"])
plt.title("Iris — Accuracy by Distance Metric (k=5)")
plt.xlabel("Metric")
plt.ylabel("Accuracy")
plt.tight_layout()
plt.show()
```


## 5) Sweep k for Selected Metric

We will sweep **k = 1..20** using **Euclidean** distance and visualize accuracy.


```python
# k sweep for Euclidean
k_values = list(range(1, 21))
acc_k = []

for k in k_values:
    model = KNeighborsClassifier(n_neighbors=k, metric='euclidean')
    model.fit(X_train_s, y_train)
    y_hat = model.predict(X_test_s)
    acc_k.append(accuracy_score(y_test, y_hat))

best_idx = int(np.argmax(acc_k))
best_k = k_values[best_idx]
best_acc = acc_k[best_idx]
print("Iris — Best k:", best_k, "with accuracy:", round(best_acc, 4))

plt.figure(figsize=(6,3.5))
plt.plot(k_values, acc_k, marker='o')
plt.title("Iris — Accuracy vs k (euclidean)")
plt.xlabel("k")
plt.ylabel("Accuracy")
plt.grid(True)
plt.tight_layout()
plt.show()
```
