```python
# EXAMPLE: Converting between types

# String to Integer
age_string = "25"
age_number = int(age_string)
print(f"String '{age_string}' converted to integer: {age_number}")
```

```python
# Integer to Float
whole_number = 10
decimal_number = float(whole_number)
print(f"Integer {whole_number} converted to float: {decimal_number}")
```

```python
# Number to String
temperature = 98.6
temp_string = str(temperature)
print(f"Float {temperature} converted to string: '{temp_string}'")
```

```python
# Boolean conversion
print(f"bool(1) = {bool(1)}")      # True
print(f"bool(0) = {bool(0)}")      # False
print(f"bool('') = {bool('')}")    # False (empty string)
print(f"bool('Hello') = {bool('Hello')}")  # True
```

```python
# String methods
email = "  user@example.com  "
print(f"Strip whitespace: '{email.strip()}'")
print(f"Replace: {text.replace('Python', 'Java')}")
print(f"Count 'o': {text.count('o')}")
```

```python
# EXAMPLE: If-elif-else statements

# Simple if statement
temperature = 75
if temperature > 70:
    print("It's warm outside!")
```

```python
# If-else
age = 18
if age >= 18:
    print("You can vote!")
else:
    print("Too young to vote")
```

```python
# If-elif-else
score = 85
if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
elif score >= 60:
    grade = "D"
else:
    grade = "F"
print(f"Score {score} = Grade {grade}")
```

```python
# Multiple conditions
is_weekend = True
is_sunny = True

if is_weekend and is_sunny:
    print("Perfect day for outdoor activities!")
elif is_weekend and not is_sunny:
    print("Good day for indoor activities")
else:
    print("It's a workday")
```

```python
# EXAMPLE: Complete temperature converter

def temperature_converter():
    """Convert between Celsius and Fahrenheit"""
    
    print("=== Temperature Converter ===")
    print("1. Celsius to Fahrenheit")
    print("2. Fahrenheit to Celsius")
    
    choice = input("Enter your choice (1 or 2): ")
    
    if choice == "1":
        celsius = float(input("Enter temperature in Celsius: "))
        fahrenheit = (celsius * 9/5) + 32
        print(f"{celsius}°C = {fahrenheit:.2f}°F")
        
        # Add description
        if celsius < 0:
            print("That's below freezing!")
        elif celsius < 10:
            print("That's cold!")
        elif celsius < 20:
            print("That's cool")
        elif celsius < 30:
            print("That's warm")
        else:
            print("That's hot!")
            
    elif choice == "2":
        fahrenheit = float(input("Enter temperature in Fahrenheit: "))
        celsius = (fahrenheit - 32) * 5/9
        print(f"{fahrenheit}°F = {celsius:.2f}°C")
        
        # Add description
        if fahrenheit < 32:
            print("That's below freezing!")
        elif fahrenheit < 50:
            print("That's cold!")
        elif fahrenheit < 68:
            print("That's cool")
        elif fahrenheit < 86:
            print("That's warm")
        else:
            print("That's hot!")
    else:
        print("Invalid choice!")

# Run the converter (uncomment to test)
# temperature_converter()
```
