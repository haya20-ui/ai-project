```python
import pandas as pd

# 1. Define the path
path = r'C:\Users\Admin\Downloads\ECommerce dataset\data.csv' 

# 2. READ the file
df = pd.read_csv(path, encoding='ISO-8859-1') 

# 3.check the shape
print(df.shape)
```

```python
# 1. Remove exact duplicates
df.drop_duplicates(inplace=True)

# 2. Handle Missing Values
df['Description'] = df['Description'].fillna('Unknown')
df['CustomerID'] = df['CustomerID'].fillna(0).astype(int) # 0 = Guest

# 3. Handle Negative Values
# We will keep only positive prices (removing those weird accounting errors)
df = df[df['UnitPrice'] > 0]

# 4. Create a new column to separate Sales from Returns
import numpy as np
df['Status'] = np.where(df['Quantity'] > 0, 'Completed', 'Cancelled')

# 5. Create a Total_Amount column (Quantity * Price)
# We use .abs() for Quantity so even cancellations show the "Value" lost
df['Total_Price'] = df['Quantity'].abs() * df['UnitPrice']

print("Cleaning Complete!")
print(f"Rows remaining: {len(df)}")
print(df.head())
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

print("="*60)
print("RFM ANALYSIS".center(60))
print("="*60)

# Prepare data
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
snapshot = df['InvoiceDate'].max() + pd.Timedelta(days=1)

# RFM table
rfm = df.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (snapshot - x.max()).days,
    'InvoiceNo': 'nunique',
    'Total_Price': 'sum'
}).rename(columns={'InvoiceDate': 'Recency', 'InvoiceNo': 'Frequency', 'Total_Price': 'Monetary'})

rfm = rfm[rfm['Monetary'] > 0]

# RFM scores
rfm['R'] = pd.qcut(rfm['Recency'], 5, labels=[5,4,3,2,1])
rfm['F'] = pd.qcut(rfm['Frequency'].rank(method='first'), 5, labels=[1,2,3,4,5])
rfm['M'] = pd.qcut(rfm['Monetary'], 5, labels=[1,2,3,4,5])
rfm['Total'] = rfm[['R','F','M']].sum(axis=1).astype(int)

# Segments
rfm['Segment'] = ' Lost'
rfm.loc[rfm['Total'] >= 5, 'Segment'] = ' At Risk'
rfm.loc[rfm['Total'] >= 7, 'Segment'] = ' Potential'
rfm.loc[rfm['Total'] >= 10, 'Segment'] = ' Loyal'
rfm.loc[rfm['Total'] >= 13, 'Segment'] = ' Champions'

# Summary
summary = rfm.groupby('Segment').agg(
    Customers=('Monetary', 'count'),
    Revenue=('Monetary', 'sum'),
    Avg_Spend=('Monetary', 'mean'),
    Avg_Recency=('Recency', 'mean'),
    Avg_Freq=('Frequency', 'mean')
).round(1)

summary['Revenue_Share'] = (summary['Revenue'] / summary['Revenue'].sum() * 100).round(1)
summary['Customer_Share'] = (summary['Customers'] / len(rfm) * 100).round(1)
summary = summary.reindex([' Champions', ' Loyal', '📈Potential', ' At Risk', ' Lost'])

print("\n CUSTOMER SEGMENT SUMMARY")
print("-" * 60)
print(summary[['Customers', 'Customer_Share', 'Revenue', 'Revenue_Share', 'Avg_Spend']].to_string())

# Insights
c = summary.loc[' Champions']
print(f"\n Top {c['Customer_Share']:.1f}% generate {c['Revenue_Share']:.1f}% of revenue")
print(f" At risk: {summary.loc[' At Risk']['Customers'] + summary.loc[' Lost']['Customers']:.0f} customers")

# SAME VISUALIZATIONS - 4 PLOTS
fig, axes = plt.subplots(2, 2, figsize=(15, 12))
colors = ['gold', 'silver', 'lightgreen', 'orange', 'lightcoral']

# 1. Pie chart - Segment distribution
seg_counts = rfm['Segment'].value_counts()
axes[0,0].pie(seg_counts.values, labels=seg_counts.index, autopct='%1.1f%%', 
              colors=colors, startangle=90)
axes[0,0].set_title('Customer Segments Distribution', fontweight='bold')

# 2. Horizontal bar - Revenue by segment
rev_by_seg = rfm.groupby('Segment')['Monetary'].sum().reindex(seg_counts.index)
bars = axes[0,1].barh(rev_by_seg.index, rev_by_seg.values, color=colors)
axes[0,1].set_xlabel('Revenue ($)')
axes[0,1].set_title('Revenue by Segment', fontweight='bold')
for bar, val in zip(bars, rev_by_seg.values):
    axes[0,1].text(val, bar.get_y() + bar.get_height()/2, f'${val:,.0f}', 
                   ha='left', va='center', fontweight='bold')

# 3. Bar chart - Average spend by segment
avg_spend = rfm.groupby('Segment')['Monetary'].mean().reindex(seg_counts.index)
axes[1,0].bar(avg_spend.index, avg_spend.values, color=colors)
axes[1,0].set_ylabel('Average Spend ($)')
axes[1,0].set_title('Average Customer Value by Segment', fontweight='bold')
axes[1,0].tick_params(axis='x', rotation=45)

# 4. Scatter plot - Recency vs Frequency
scatter = axes[1,1].scatter(rfm['Recency'], rfm['Frequency'], 
                           c=pd.Categorical(rfm['Segment']).codes, 
                           cmap='viridis', alpha=0.6, s=50)
axes[1,1].set_xlabel('Recency (days)')
axes[1,1].set_ylabel('Frequency')
axes[1,1].set_title('Customer Behavior: Recency vs Frequency', fontweight='bold')
legend = [plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=colors[i], 
                     label=seg_counts.index[i], markersize=10) for i in range(len(seg_counts))]
axes[1,1].legend(handles=legend, loc='upper right')

plt.tight_layout()
plt.show()

# Additional Metrics
print("\n ADDITIONAL METRICS")
print("-" * 60)
print(f"Total Customers: {len(rfm):,}")
print(f"Total Revenue: ${rfm['Monetary'].sum():,.2f}")
print(f"Avg Order Value: ${rfm['Monetary'].sum()/rfm['Frequency'].sum():,.2f}")
print(f"Customer Lifetime Value: ${rfm['Monetary'].mean():,.2f}")

print("\n RFM Complete - customer_segments.csv saved")
rfm.to_csv('customer_segments.csv')
```

```python
# =============================================================================
# PRODUCT ANALYSIS - PART 3: PRICE POINTS
# =============================================================================

def analyze_price_points(df):
    """
    Find which price ranges sell best
    """
    print("\n" + "="*60)
    print("PRICE POINT ANALYSIS".center(60))
    print("="*60)
    
    sales = df[df['Quantity'] > 0].copy()
    
    # Create price segments
    bins = [0, 5, 10, 20, 50, 100, 500, 1000, 10000]
    labels = ['<$5', '$5-10', '$10-20', '$20-50', '$50-100', '$100-500', '$500-1000', '>$1000']
    
    sales['Price_Range'] = pd.cut(sales['UnitPrice'], bins=bins, labels=labels)
    
    # Analyze each price range
    price_analysis = sales.groupby('Price_Range').agg({
        'Total_Price': 'sum',
        'Quantity': 'sum',
        'InvoiceNo': 'nunique'
    }).round(2)
    
    price_analysis.columns = ['Revenue', 'Units_Sold', 'Orders']
    price_analysis['Revenue_Share'] = (price_analysis['Revenue'] / price_analysis['Revenue'].sum() * 100).round(1)
    price_analysis = price_analysis.sort_index()
    
    print("\n💰 REVENUE BY PRICE RANGE:")
    print("-" * 50)
    for idx, row in price_analysis.iterrows():
        if not pd.isna(idx):
            print(f"{idx}:")
            print(f"  Revenue: ${row['Revenue']:,.0f} ({row['Revenue_Share']}%)")
            print(f"  Units: {row['Units_Sold']:,.0f} | Orders: {row['Orders']:,.0f}")
    
    # Best price point
    best = price_analysis['Revenue'].idxmax()
    print(f"\n🏆 BEST PRICE POINT: {best}")
    print(f"   Generates {price_analysis.loc[best, 'Revenue_Share']}% of total revenue")
    
    # Visualization
    plt.figure(figsize=(10, 5))
    price_analysis['Revenue'].plot(kind='bar', color='green', edgecolor='black')
    plt.title('Revenue by Price Range', fontweight='bold', fontsize=14)
    plt.xlabel('Price Range')
    plt.ylabel('Revenue ($)')
    plt.xticks(rotation=45)
    plt.grid(axis='y', alpha=0.3)
    plt.tight_layout()
    plt.show()
    
    return price_analysis

# Run Part 3
price_results = analyze_price_points(df)
```

```python
# =============================================================================
# PRODUCT ANALYSIS - PART 4: PRODUCT COMBINATIONS
# =============================================================================

def find_product_pairs(df, top_n=10):
    """
    Find products frequently bought together
    """
    print("\n" + "="*60)
    print("PRODUCT COMBINATIONS".center(60))
    print("="*60)
    
    sales = df[df['Quantity'] > 0]
    
    # Get products per invoice
    invoice_groups = sales.groupby('InvoiceNo')['StockCode'].apply(list)
    
    # Find pairs
    from collections import Counter
    pairs = []
    
    for products in invoice_groups:
        if len(products) > 1:
            for i in range(len(products)):
                for j in range(i+1, len(products)):
                    # Sort to ensure consistent ordering
                    pair = tuple(sorted([products[i], products[j]]))
                    pairs.append(pair)
    
    # Count frequencies
    pair_counts = Counter(pairs).most_common(top_n)
    
    print("\n🛒 TOP PRODUCTS BOUGHT TOGETHER:")
    print("-" * 50)
    
    for i, (pair, count) in enumerate(pair_counts, 1):
        # Get descriptions
        desc1 = sales[sales['StockCode'] == pair[0]]['Description'].iloc[0] if len(sales[sales['StockCode'] == pair[0]]) > 0 else pair[0]
        desc2 = sales[sales['StockCode'] == pair[1]]['Description'].iloc[0] if len(sales[sales['StockCode'] == pair[1]]) > 0 else pair[1]
        
        print(f"\n{i}. {desc1[:40]}...")
        print(f"   + {desc2[:40]}...")
        print(f"   Bought together in {count} orders")
    
    return pair_counts

# Run Part 4
product_pairs = find_product_pairs(df)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose
import warnings
warnings.filterwarnings('ignore')

# Set style
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")

# =============================================================================
# TIME SERIES ANALYSIS - PART 1: PREPARE DATA
# =============================================================================

def prepare_time_series(df):
    """
    Prepare data for time series analysis
    """
    print("="*60)
    print("TIME SERIES ANALYSIS".center(60))
    print("="*60)
    
    # Convert to datetime
    df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
    
    # Daily revenue
    daily_revenue = df.groupby(df['InvoiceDate'].dt.date)['Total_Price'].sum().reset_index()
    daily_revenue.columns = ['Date', 'Revenue']
    daily_revenue['Date'] = pd.to_datetime(daily_revenue['Date'])
    daily_revenue = daily_revenue.sort_values('Date')
    
    # Weekly revenue
    weekly_revenue = df.groupby(df['InvoiceDate'].dt.to_period('W'))['Total_Price'].sum().reset_index()
    weekly_revenue.columns = ['Week', 'Revenue']
    weekly_revenue['Week'] = weekly_revenue['Week'].astype(str)
    
    # Monthly revenue
    monthly_revenue = df.groupby(df['InvoiceDate'].dt.to_period('M'))['Total_Price'].sum().reset_index()
    monthly_revenue.columns = ['Month', 'Revenue']
    monthly_revenue['Month'] = monthly_revenue['Month'].astype(str)
    
    print(f"\n📊 Data Prepared:")
    print(f"   • Daily data: {len(daily_revenue)} days")
    print(f"   • Weekly data: {len(weekly_revenue)} weeks")
    print(f"   • Monthly data: {len(monthly_revenue)} months")
    
    return daily_revenue, weekly_revenue, monthly_revenue

# Run Part 1
daily, weekly, monthly = prepare_time_series(df)
```

```python
# =============================================================================
# TIME SERIES ANALYSIS - PART 2: DAILY PATTERNS
# =============================================================================

def analyze_daily_patterns(df):
    """
    Analyze daily, weekly, and hourly patterns
    """
    print("\n" + "="*60)
    print("DAILY & WEEKLY PATTERNS".center(60))
    print("="*60)
    
    # Add time features
    df['Day_of_Week'] = df['InvoiceDate'].dt.day_name()
    df['Hour'] = df['InvoiceDate'].dt.hour
    df['Month'] = df['InvoiceDate'].dt.month_name()
    
    # Revenue by day of week
    day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
    daily_pattern = df.groupby('Day_of_Week')['Total_Price'].agg(['sum', 'mean', 'count']).round(2)
    daily_pattern = daily_pattern.reindex(day_order)
    daily_pattern.columns = ['Total_Revenue', 'Avg_Transaction', 'Transaction_Count']
    daily_pattern['Share'] = (daily_pattern['Total_Revenue'] / daily_pattern['Total_Revenue'].sum() * 100).round(1)
    
    print("\n📅 REVENUE BY DAY OF WEEK:")
    print("-" * 50)
    for day in day_order:
        if day in daily_pattern.index:
            rev = daily_pattern.loc[day, 'Total_Revenue']
            share = daily_pattern.loc[day, 'Share']
            avg = daily_pattern.loc[day, 'Avg_Transaction']
            print(f"{day:10}: ${rev:>10,.0f} ({share:>4}%) | Avg Order: ${avg:>7,.2f}")
    
    # Revenue by hour
    hourly_pattern = df.groupby('Hour')['Total_Price'].agg(['sum', 'mean', 'count']).round(2)
    hourly_pattern.columns = ['Total_Revenue', 'Avg_Transaction', 'Transaction_Count']
    
    print("\n⏰ REVENUE BY HOUR:")
    print("-" * 50)
    peak_hour = hourly_pattern['Total_Revenue'].idxmax()
    peak_value = hourly_pattern.loc[peak_hour, 'Total_Revenue']
    print(f"Peak hour: {peak_hour}:00 (${peak_value:,.0f} revenue)")
    
    top_hours = hourly_pattern.nlargest(3, 'Total_Revenue')
    for hour in top_hours.index:
        print(f"  {hour:02d}:00: ${top_hours.loc[hour, 'Total_Revenue']:,.0f}")
    
    # Visualizations
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))
    
    # Day of week bar chart
    axes[0].bar(daily_pattern.index, daily_pattern['Total_Revenue'], color='skyblue', edgecolor='black')
    axes[0].set_title('Revenue by Day of Week', fontweight='bold', fontsize=14)
    axes[0].set_xlabel('Day')
    axes[0].set_ylabel('Revenue ($)')
    axes[0].tick_params(axis='x', rotation=45)
    
    # Hourly line chart
    axes[1].plot(hourly_pattern.index, hourly_pattern['Total_Revenue'], marker='o', linewidth=2, markersize=6)
    axes[1].set_title('Revenue by Hour of Day', fontweight='bold', fontsize=14)
    axes[1].set_xlabel('Hour')
    axes[1].set_ylabel('Revenue ($)')
    axes[1].grid(True, alpha=0.3)
    axes[1].axvspan(9, 17, alpha=0.2, color='green', label='Business Hours')
    axes[1].legend()
    
    plt.tight_layout()
    plt.show()
    
    return daily_pattern, hourly_pattern

# Run Part 2
daily_pattern, hourly_pattern = analyze_daily_patterns(df)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")

# =============================================================================
# TRANSACTION ANALYSIS - PART 3: CUSTOMER BUYING PATTERNS (FIXED)
# =============================================================================

def customer_buying_patterns(df):
    """
    Analyze how customers buy over time
    """
    print("\n" + "="*60)
    print("CUSTOMER BUYING PATTERNS".center(60))
    print("="*60)
    
    sales = df[df['Quantity'] > 0].copy()
    
    # Transactions per customer
    customer_frequency = sales.groupby('CustomerID')['InvoiceNo'].nunique()
    
    # Calculate counts
    one_time = len(customer_frequency[customer_frequency == 1])
    two_three = len(customer_frequency[(customer_frequency >= 2) & (customer_frequency <= 3)])
    four_ten = len(customer_frequency[(customer_frequency > 3) & (customer_frequency <= 10)])
    ten_plus = len(customer_frequency[customer_frequency > 10])
    total_customers = len(customer_frequency)
    
    print("\n👥 TRANSACTIONS PER CUSTOMER:")
    print("-" * 50)
    print(f"• One-time buyers: {one_time} customers ({one_time/total_customers*100:.1f}%)")
    print(f"• 2-3 times: {two_three} customers ({two_three/total_customers*100:.1f}%)")
    print(f"• 4-10 times: {four_ten} customers ({four_ten/total_customers*100:.1f}%)")
    print(f"• 10+ times: {ten_plus} customers ({ten_plus/total_customers*100:.1f}%)")
    
    # Calculate repeat rate (FIXED: removed .round())
    repeat_customers = len(customer_frequency[customer_frequency > 1])
    repeat_rate = (repeat_customers / total_customers * 100)
    
    print(f"\n🔄 Repeat Customer Rate: {repeat_rate:.1f}%")
    print(f"   • {repeat_customers} customers bought more than once")
    print(f"   • {one_time} customers bought only once")
    
    # Time between purchases
    customer_dates = sales.groupby('CustomerID')['InvoiceDate'].apply(list)
    
    purchase_gaps = []
    for cust, dates in customer_dates.items():
        if len(dates) > 1:
            dates = sorted(dates)
            for i in range(1, len(dates)):
                gap = (dates[i] - dates[i-1]).days
                purchase_gaps.append(gap)
    
    if purchase_gaps:
        avg_gap = np.mean(purchase_gaps)
        median_gap = np.median(purchase_gaps)
        print(f"\n⏱️  TIME BETWEEN PURCHASES:")
        print(f"   • Average: {avg_gap:.0f} days")
        print(f"   • Median: {median_gap:.0f} days")
        print(f"   • Min gap: {min(purchase_gaps)} days")
        print(f"   • Max gap: {max(purchase_gaps)} days")
    
    # Key insight
    print("\n" + "="*60)
    print("KEY INSIGHT".center(60))
    print("="*60)
    
    if repeat_rate < 30:
        print(f"\n⚠️  Low repeat rate ({repeat_rate:.1f}%): Focus on retention")
        print("   • Implement loyalty program")
        print("   • Send post-purchase emails")
        print("   • Offer repeat customer discounts")
    elif repeat_rate < 50:
        print(f"\n📊 Moderate repeat rate ({repeat_rate:.1f}%): Room to improve")
        print("   • Create referral program")
        print("   • Target one-time buyers with reactivation")
    else:
        print(f"\n🌟 Strong repeat rate ({repeat_rate:.1f}%): Nurture loyalists")
        print("   • VIP program for frequent buyers")
        print("   • Early access to new products")
    
    return customer_frequency, purchase_gaps

# Run Part 3
customer_frequency, purchase_gaps = customer_buying_patterns(df)

print("\n✅ Part 3 Complete - Customer frequency data ready")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")

print("="*60)
print("GEOGRAPHIC ANALYSIS - PART 1".center(60))
print("="*60)

def country_performance(df):
    """Analyze revenue and customers by country"""
    
    # Sales only
    sales = df[df['Quantity'] > 0].copy()
    
    # Country metrics
    country_stats = sales.groupby('Country').agg({
        'Total_Price': ['sum', 'mean'],
        'CustomerID': 'nunique',
        'InvoiceNo': 'nunique'
    }).round(2)
    
    country_stats.columns = ['Revenue', 'Avg_Order', 'Customers', 'Transactions']
    country_stats = country_stats.sort_values('Revenue', ascending=False)
    
    # Add percentages
    total_revenue = country_stats['Revenue'].sum()
    country_stats['Revenue_Share'] = (country_stats['Revenue'] / total_revenue * 100).round(1)
    
    # Top 10 countries
    top_10 = country_stats.head(10).copy()
    
    print("\n TOP 10 COUNTRIES BY REVENUE:")
    print("-" * 60)
    print(f"{'Country':<15} {'Revenue':>12} {'Share':>6} {'Customers':>10}")
    print("-" * 60)
    
    for country, row in top_10.iterrows():
        print(f"{country[:15]:<15} ${row['Revenue']:>10,.0f} {row['Revenue_Share']:>5}% {row['Customers']:>8.0f}")
    
    # Market concentration
    uk_share = country_stats.loc['United Kingdom', 'Revenue_Share'] if 'United Kingdom' in country_stats.index else 0
    top3_share = country_stats.head(3)['Revenue_Share'].sum()
    top1_share = country_stats.head(1)['Revenue_Share'].values[0]
    
    print(f"\n MARKET CONCENTRATION:")
    print(f"   • #1 country: {top1_share:.1f}% of revenue")
    print(f"   • Top 3 countries: {top3_share:.1f}% of revenue")
    print(f"   • UK: {uk_share:.1f}% of revenue")
    
    return country_stats

# Run Part 1
country_stats = country_performance(df)
```

```python
print("\n" + "="*60)
print("GEOGRAPHIC ANALYSIS - PART 2".center(60))
print("="*60)

def customer_metrics_by_country(df):
    """Analyze customer behavior by country"""
    
    sales = df[df['Quantity'] > 0].copy()
    
    # Customers per country
    customers_by_country = sales.groupby('Country')['CustomerID'].nunique().sort_values(ascending=False)
    
    print("\n CUSTOMERS BY COUNTRY:")
    print("-" * 50)
    print(f"{'Country':<15} {'Customers':>10} {'% of Total':>10}")
    print("-" * 50)
    
    total_customers = customers_by_country.sum()
    for country, count in customers_by_country.head(8).items():
        pct = (count / total_customers * 100).round(1)
        print(f"{country[:15]:<15} {count:>10.0f} {pct:>9}%")
    
    # Repeat rate by country
    print("\n REPEAT CUSTOMER RATE BY COUNTRY:")
    print("-" * 50)
    
    repeat_rates = {}
    for country in customers_by_country.head(8).index:
        country_customers = sales[sales['Country'] == country].groupby('CustomerID')['InvoiceNo'].nunique()
        repeat = len(country_customers[country_customers > 1])
        total = len(country_customers)
        rate = (repeat / total * 100) if total > 0 else 0
        repeat_rates[country] = round(rate, 1)
        
        print(f"{country[:15]:<15} {rate:>5}% repeat ({repeat}/{total})")
    
    return customers_by_country, repeat_rates

# Run Part 2
customers_by_country, repeat_rates = customer_metrics_by_country(df)
```

```python
print("\n" + "="*60)
print("GEOGRAPHIC ANALYSIS - PART 3".center(60))
print("="*60)

def order_value_by_country(df):
    """Analyze average order value by country"""
    
    sales = df[df['Quantity'] > 0].copy()
    
    # AOV by country
    aov = sales.groupby('Country')['Total_Price'].mean().round(2).sort_values(ascending=False)
    
    print("\n AVERAGE ORDER VALUE BY COUNTRY:")
    print("-" * 50)
    print(f"{'Country':<15} {'Avg Order':>12}")
    print("-" * 50)
    
    for country, value in aov.head(10).items():
        print(f"{country[:15]:<15} ${value:>10.2f}")
    
    # UK comparison
    uk_aov = aov.get('United Kingdom', 0)
    print(f"\n UK Average: ${uk_aov:.2f}")
    
    # Countries with higher AOV than UK
    higher = []
    for country, value in aov.items():
        if value > uk_aov and country != 'United Kingdom':
            higher.append(f"{country} (${value:.0f})")
    
    if higher:
        print(f"\n Countries with HIGHER AOV than UK:")
        for h in higher[:5]:
            print(f"   • {h}")
    
    return aov

# Run Part 3
aov = order_value_by_country(df)
```

```python
print("\n" + "="*60)
print("GEOGRAPHIC ANALYSIS - PART 4".center(60))
print("="*60)

def returns_by_country(df):
    """Analyze return rates by country"""
    
    returns = df[df['Quantity'] < 0].copy()
    sales = df[df['Quantity'] > 0].copy()
    
    if len(returns) == 0:
        print("\n No returns found in data")
        return None
    
    # Return value by country
    return_value = abs(returns.groupby('Country')['Total_Price'].sum())
    sales_value = sales.groupby('Country')['Total_Price'].sum()
    
    # Calculate return rates
    return_rates = {}
    for country in sales_value.index:
        if country in return_value.index:
            rate = (return_value[country] / sales_value[country] * 100).round(2)
            return_rates[country] = rate
    
    # Sort and display
    sorted_rates = dict(sorted(return_rates.items(), key=lambda x: x[1], reverse=True))
    
    print("\n RETURN RATE BY COUNTRY:")
    print("-" * 50)
    print(f"{'Country':<15} {'Return Rate':>12}")
    print("-" * 50)
    
    for country, rate in list(sorted_rates.items())[:8]:
        print(f"{country[:15]:<15} {rate:>10}%")
    
    # Average return rate
    avg_return = (returns['Total_Price'].sum() / sales['Total_Price'].sum() * 100).round(2)
    print(f"\n Overall return rate: {avg_return}%")
    
    return return_rates

# Run Part 4
return_rates = returns_by_country(df)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print("="*60)
print("STATISTICAL ANALYSIS - PART 1".center(60))
print("="*60)

# Select numerical columns
num_cols = ['Quantity', 'UnitPrice', 'Total_Price']
corr_data = df[num_cols].copy()

# Calculate correlations
corr_matrix = corr_data.corr().round(3)

print("\n CORRELATION MATRIX:")
print("-" * 40)
print(corr_matrix)

print("\n KEY CORRELATIONS:")
for i in range(len(corr_matrix.columns)):
    for j in range(i+1, len(corr_matrix.columns)):
        col1 = corr_matrix.columns[i]
        col2 = corr_matrix.columns[j]
        val = corr_matrix.iloc[i, j]
        strength = "Strong" if abs(val) > 0.5 else "Moderate" if abs(val) > 0.3 else "Weak"
        print(f"• {col1} vs {col2}: {val:.3f} ({strength})")

# Heatmap
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0, square=True)
plt.title('Correlation Matrix', fontweight='bold')
plt.show()
```

```python
print("\n" + "="*60)
print("STATISTICAL ANALYSIS - PART 2".center(60))
print("="*60)

def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = data[(data[column] < lower) | (data[column] > upper)]
    return outliers, lower, upper

# Check each numerical column
for col in ['Quantity', 'UnitPrice', 'Total_Price']:
    data_clean = df[df[col].notna() & (df[col] > 0)]
    outliers, lower, upper = detect_outliers_iqr(data_clean, col)
    
    print(f"\n OUTLIERS IN {col}:")
    print("-" * 40)
    print(f"Normal range: [{lower:.2f}, {upper:.2f}]")
    print(f"Outliers: {len(outliers)} ({len(outliers)/len(data_clean)*100:.1f}%)")
    print(f"Min: {data_clean[col].min():.2f} | Max: {data_clean[col].max():.2f}")
    
    if len(outliers) > 0:
        print(f"Top 5 outliers: {outliers[col].nlargest(5).values}")

# Box plots
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
for i, col in enumerate(['Quantity', 'UnitPrice', 'Total_Price']):
    axes[i].boxplot(df[df[col] > 0][col])
    axes[i].set_title(f'{col} Outliers')
    axes[i].set_ylabel(col)
plt.tight_layout()
plt.show()
```

```python
print("\n" + "="*60)
print("STATISTICAL ANALYSIS - PART 3".center(60))
print("="*60)

# Distribution stats for numerical columns
for col in ['Quantity', 'UnitPrice', 'Total_Price']:
    data = df[df[col] > 0][col]
    
    print(f"\n📊 {col.upper()} DISTRIBUTION:")
    print("-" * 40)
    print(f"Mean:     {data.mean():.2f}")
    print(f"Median:   {data.median():.2f}")
    print(f"Mode:     {data.mode()[0]:.2f}")
    print(f"Std Dev:  {data.std():.2f}")
    print(f"Variance: {data.var():.2f}")
    print(f"Skewness: {data.skew():.2f}")
    print(f"Kurtosis: {data.kurtosis():.2f}")
    
    # Percentiles
    p = [0, 25, 50, 75, 90, 95, 99, 100]
    percentiles = np.percentile(data, p).round(2)
    print("\nPercentiles:")
    for pi, val in zip(p, percentiles):
        print(f"  {pi:3}th: {val:8.2f}")

# Histograms
fig, axes = plt.subplots(1, 3, figsize=(15, 4))
for i, col in enumerate(['Quantity', 'UnitPrice', 'Total_Price']):
    data = df[df[col] > 0][col].clip(upper=data.quantile(0.99))
    axes[i].hist(data, bins=30, edgecolor='black')
    axes[i].set_title(f'{col} Distribution')
    axes[i].axvline(data.mean(), color='red', linestyle='--', label=f'Mean: {data.mean():.1f}')
    axes[i].axvline(data.median(), color='green', linestyle='--', label=f'Median: {data.median():.1f}')
    axes[i].legend()
plt.tight_layout()
plt.show()
```

```python
# TOTAL REVENUE & AOV
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

print("="*60)
print("KEY METRICS - PART 1".center(60))
print("="*60)

# Sales only (positive quantity)
sales = df[df['Quantity'] > 0].copy()

# Total Revenue
total_revenue = sales['Total_Price'].sum()
print(f"\n TOTAL REVENUE: ${total_revenue:,.2f}")

# Average Order Value
total_transactions = sales['InvoiceNo'].nunique()
aov = total_revenue / total_transactions
print(f" TOTAL TRANSACTIONS: {total_transactions:,}")
print(f" AVERAGE ORDER VALUE: ${aov:.2f}")

# Quick visualization
plt.figure(figsize=(8,4))
plt.bar(['Total Revenue'], [total_revenue], color='green')
plt.title(f'Total Revenue: ${total_revenue:,.0f}')
plt.ylabel('Revenue ($)')
plt.show()
```

```python
print("\n" + "="*60)
print("KEY METRICS - PART 4 (FIXED)".center(60))
print("="*60)

# Revenue per customer
customer_value = sales.groupby('CustomerID')['Total_Price'].sum()

avg_rpc = customer_value.mean()
median_rpc = customer_value.median()

print(f"\n💰 REVENUE PER CUSTOMER (RPC):")
print(f"• Average: ${avg_rpc:,.2f}")
print(f"• Median: ${median_rpc:,.2f}")
print(f"• Min: ${customer_value.min():.2f}")
print(f"• Max: ${customer_value.max():,.2f}")

# Customer value segments
segments = ['Under $100', '$100-500', '$500-1k', '$1k-5k', 'Over $5k']
counts = [
    len(customer_value[customer_value < 100]),
    len(customer_value[(customer_value >= 100) & (customer_value < 500)]),
    len(customer_value[(customer_value >= 500) & (customer_value < 1000)]),
    len(customer_value[(customer_value >= 1000) & (customer_value < 5000)]),
    len(customer_value[customer_value >= 5000])
]

print(f"\n📊 CUSTOMER VALUE BREAKDOWN:")
for seg, cnt in zip(segments, counts):
    pct = (cnt / len(customer_value) * 100) if len(customer_value) > 0 else 0
    print(f"  {seg:12}: {cnt:4} customers ({pct:.1f}%)")

# Bar chart
plt.figure(figsize=(8,4))
plt.bar(segments, counts, color='purple', edgecolor='black')
plt.title('Customer Value Segments')
plt.xlabel('Customer Lifetime Value')
plt.ylabel('Number of Customers')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
print("\n" + "="*60)
print("KEY METRICS - PART 5".center(60))
print("="*60)

# Monthly revenue
sales['Month'] = pd.to_datetime(sales['InvoiceDate']).dt.month_name()
monthly = sales.groupby('Month')['Total_Price'].sum()

# Calculate seasonal index (100 = average)
avg_monthly = monthly.mean()
seasonal_index = (monthly / avg_monthly * 100).round(1)

# Sort by calendar order
month_order = ['January', 'February', 'March', 'April', 'May', 'June',
               'July', 'August', 'September', 'October', 'November', 'December']
seasonal_index = seasonal_index.reindex(month_order)

print("\n📅 SEASONAL INDICES (100 = average month):")
print("-" * 50)
for month, idx in seasonal_index.items():
    if not pd.isna(idx):
        bar = '█' * int(idx/5)
        print(f"{month:10}: {idx:5} {bar}")

# Best and worst
best_month = seasonal_index.idxmax()
best_val = seasonal_index.max()
worst_month = seasonal_index.idxmin()
worst_val = seasonal_index.min()

print(f"\n PEAK MONTH: {best_month} ({best_val}% of average)")
print(f" SLOW MONTH: {worst_month} ({worst_val}% of average)")
print(f" Q4 (Oct-Dec): {(seasonal_index['October'] + seasonal_index['November'] + seasonal_index['December'])/3:.0f}% of average")

# Line chart
plt.figure(figsize=(10,4))
plt.plot(seasonal_index.index, seasonal_index.values, marker='o', linewidth=2)
plt.title('Seasonal Indices by Month')
plt.xlabel('Month')
plt.ylabel('Index (100 = average)')
plt.axhline(y=100, color='red', linestyle='--', label='Average')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

```
