# Exploratory Data Analysis (EDA)


**Our Mission:**
1.  **Find Mistakes:** Like duplicate students or impossible ages.
2.  **Fix Missing Data:** Decide when to delete and when to guess (impute).
3.  **Analyze Patterns:** See what affects student grades.

---

###  Step 1: Get Your Tools
We need to load our Python libraries (tools). Run the cell below.

```python
import pandas as pd          
import seaborn as sns          
import matplotlib.pyplot as plt 
import numpy as np              
```

### Step 2: Create the "Messy" Data
Usually, you load a file (like Excel). Today, we will **generate** a fake dataset right here so you have specific problems to fix.

**The Dataset Contains:**
* **Duplicates:** Some students were entered twice.
* **Outliers:** One student is 500 years old (Impossible!).
* **Missing Data (Small):** `Math_Score` has a few blanks.
* **Missing Data (Huge):** `Locker_Number` is missing for almost everyone.

```python
# ---------------------------------------------------------------
# GENERATING PRACTICE DATA (Just run this cell!)
# ---------------------------------------------------------------
np.random.seed(42)


data = {
    'Student_ID': range(101, 201),
    'Age': np.random.randint(15, 19, 100),  # Normal ages 15-18
    'Math_Score': np.random.normal(70, 10, 100).round(1),
    'Reading_Score': np.random.normal(75, 10, 100).round(1),
    'Study_Hours': np.random.choice([1, 2, 3, 4, 5], 100),
    'Favorite_Subject': np.random.choice(['Math', 'Science', 'History', 'Art'], 100),
    'Locker_Number': [np.nan] * 90 + list(range(1, 11)) # 90% Missing!
}

df = pd.DataFrame(data)


# 1. Add Duplicates (Student 101 appears 3 times)
df = pd.concat([df, df.iloc[[0, 0]]], ignore_index=True)

# 2. Add Outlier (Impossible Age)
df.loc[5, 'Age'] = 500 

# 3. Add Missing Values (Small amount in Math_Score)
df.loc[10:14, 'Math_Score'] = np.nan
```

---

##  Phase 1: The Sanity Check
Before analysis, we just look at the data structure.

**What to do:**
1.  Use `df.head()` to see the top rows.
2.  Use `df.info()` to check data types and counts.

```python
# Look at the first 5 rows
print(df.head())

# Check data types and non-null counts

df.info()
```

---

## Phase 2: Cleaning - Duplicates

**Analysis:**
Check if any rows are repeated exactly. 

**Action Rule:**
* If it's an **exact duplicate**, it is usually a data entry error. **Action: Delete it.**

```python
# Check for duplicates
num_duplicates = df.duplicated().sum()
print(f" Found {num_duplicates} duplicate rows.")

if num_duplicates > 0:
    print(" Action: Removing duplicates...")
    df = df.drop_duplicates()
    print(" Duplicates removed.")
else:
    print(" No duplicates found.")
```

---

## Phase 3: Cleaning - Missing Values (Nulls)

**Analysis:**
Find which columns have empty cells (`NaN`).

**Action Rules:**
1.  **Missing A Lot (>50%):** The column is mostly empty. It teaches us nothing. **Action: Drop the Column.**
2.  **Missing A Little (<5%):** Just a few cells. **Action: Remove it / Fill it (Impute).**

```python
# See missing values per column
print("Missing Values Report:")
print(df.isnull().sum())
```

###  Taking Action on Nulls
Look at the report above.
* `Locker_Number`: Missing 90 times! -> **Drop Column**.
* `Math_Score`: Missing 5 times. -> **Fill / Remove**.

```python
# 1. Handling the BIG missing column
df = df.drop(columns=['Locker_Number'])

# 2. Handling the SMALL missing values
df = df.dropna()

```

---

##  Phase 4: Univariate Analysis (Numerical)
**Goal:** Understand the distribution of numbers.

**We look for:**
1.  **Outliers:** Values that are impossibly high or low.
2.  **Shape:** Is it a bell curve (Normal)?

**Tools:** `Histogram` (Shape) and `Boxplot` (Outliers).

```python
# Let's analyze AGE
column = 'Age'

print(f"Statistics for {column}:")
print(df[column].describe())

plt.figure()

# Graph 1: Histogram
sns.histplot(df[column], kde=True)
plt.title(f"Histogram of {column}")

plt.figure()
# Graph 2: Boxplot
sns.boxplot(x=df[column])
plt.title(f"Boxplot of {column}")

plt.show()
```

### Analyzing the Result
**What do we see?**
* **Boxplot:** There is a single dot way off to the right (at 500).
* **Action:** This is an impossible age for a student. It is an error. **We must remove it.**

```python
# Filter data to keep only realistic ages (e.g., less than 100)
print(f"Before cleaning: {len(df)} rows")
df = df[df['Age'] < 100]
print(f"After cleaning: {len(df)} rows (Outlier removed)")
```

---

## Phase 5: Univariate Analysis (Categorical)
**Goal:** Count the categories.

**We look for:**
1.  **Dominance:** Is one category 99% of the data? (If so, it's not useful).
2.  **Typos:** "Math" vs "math" (We would need to fix these, but our data is clean).

```python
col_cat = 'Favorite_Subject'

print(f"Counts for {col_cat}:")
print(df[col_cat].value_counts())

plt.figure(figsize=(8, 4))
sns.countplot(x=df[col_cat])
plt.title(f"Student Preference: {col_cat}")
plt.show()
```

---

## Phase 6: Multivariate Analysis (Relationships)
**Goal:** See how variables affect each other.

### 1. Numerical vs Numerical
**Tool:** Scatter Plot.
**Question:** Does `Reading_Score` go up when `Math_Score` goes up?

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df['Math_Score'], y=df['Reading_Score'])
plt.title("Relationship: Math vs Reading Scores")
plt.show()

# Calculate Correlation (Number between -1 and 1)
corr = df['Math_Score'].corr(df['Reading_Score'])
print(f"Correlation: {corr:.2f}")
```

**Analysis:**
* If Correlation is close to **0**, there is **no relationship**.
* If Correlation is close to **1**, they strongly **increase together**.

### 2. Numerical vs Categorical
**Tool:** Boxplot.
**Question:** Do students who like "Math" actually get higher `Math_Scores`?

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['Favorite_Subject'], y=df['Math_Score'])
plt.title("Math Scores grouped by Favorite Subject")
plt.show()
```

**Analysis:**
* Look at the black line in the middle of each box (the Median).
* If the 'Math' box is higher than the 'Art' box, then liking Math is related to higher scores.
