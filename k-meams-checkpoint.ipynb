```python
# @title K-means Clustering on Random 2D Points (Using Built-in KMeans)
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

np.random.seed(42)  # Set random seed for reproducibility
# 1. Generate Random 2D Points
print("="*50)
print("STEP 1: Generating Random 2D Points")
print("="*50)

# Generate random 2D points
n_samples = 300
# Generate random points in the range [0, 10] for both x and y coordinates
X = np.random.rand(n_samples, 2) * 10 #Multiply by 10: spread points 0-10 range

print(f"Generated {n_samples} random 2D points")
print(f"Data shape: {X.shape}")
print(f"First 5 points:\n{X[:5]}")
print(f"X coordinate range: [{X[:, 0].min():.2f}, {X[:, 0].max():.2f}]")
print(f"Y coordinate range: [{X[:, 1].min():.2f}, {X[:, 1].max():.2f}]")

# Visualize the generated random points
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c='blue', s=50, alpha=0.7)
plt.title(f'Generated Random 2D Points (No Natural Clusters)', fontsize=14)
plt.xlabel('X coordinate')
plt.ylabel('Y coordinate')
plt.grid(True, alpha=0.3)
plt.xlim(0, 10)
plt.ylim(0, 10)
plt.show()

# 2. Apply K-means Clustering
print("\n" + "="*50)
print("STEP 2: Applying K-means Clustering")
print("="*50)

n_clusters = 4 # Choose number of clusters

# Create and fit KMeans model
kmeans = KMeans(n_clusters=n_clusters,
                random_state=42,
                n_init=10,
                max_iter=300)

kmeans.fit(X) # Fit the model to the data

# Get the cluster labels and centroids
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

print("K-means Results:")
print(f"Number of iterations: {kmeans.n_iter_}")
print(f"Inertia (sum of squared distances): {kmeans.inertia_:.2f}")
print(f"\nCentroids:")
for i, centroid in enumerate(centroids):
    print(f"Cluster {i+1}: [{centroid[0]:.2f}, {centroid[1]:.2f}]")

# 3. Visualize Clustering Results
print("\n" + "="*50)
print("STEP 3: Visualizing Results")
print("="*50)

# Create visualization
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Plot 1: Original random points
ax1.scatter(X[:, 0], X[:, 1], c='blue', s=50, alpha=0.7)
ax1.set_title('Original Random Points', fontsize=12)
ax1.set_xlabel('X coordinate')
ax1.set_ylabel('Y coordinate')
ax1.grid(True, alpha=0.3)
ax1.set_xlim(0, 10)
ax1.set_ylim(0, 10)

# Plot 2: K-means clustering results
scatter = ax2.scatter(X[:, 0], X[:, 1], c=labels,cmap='viridis',s=50,alpha=0.7)
ax2.scatter(centroids[:, 0], centroids[:, 1],
           marker='X', s=200, c='red', edgecolors='black', linewidths=2,
           label='Centroids')
ax2.set_title(f'K-means Clustering Results (k={n_clusters})', fontsize=12)
ax2.set_xlabel('X coordinate')
ax2.set_ylabel('Y coordinate')
ax2.legend()
ax2.grid(True, alpha=0.3)
ax2.set_xlim(0, 10)
ax2.set_ylim(0, 10)

plt.tight_layout()
plt.show()

# 4. Elbow Method to Find Optimal K
print("\n" + "="*50)
print("STEP 4: Elbow Method to Find Optimal Number of Clusters")
print("="*50)

# Calculate inertia for different values of k
k_range = range(1, 11)
inertias = []

for k in k_range:
    kmeans_test = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans_test.fit(X)
    inertias.append(kmeans_test.inertia_)

# Plot elbow curve
plt.figure(figsize=(8, 6))
plt.plot(k_range, inertias, 'bo-', linewidth=2, markersize=8)
plt.xlabel('Number of Clusters (k)', fontsize=12)
plt.ylabel('Inertia (Sum of Squared Distances)', fontsize=12)
plt.title('Elbow Method for Optimal k (Random Points)', fontsize=14)
plt.grid(True, alpha=0.3)
plt.xticks(k_range)

# Find the elbow point (where the rate of decrease slows down)
# For random data, the elbow is often less pronounced
diffs = np.diff(inertias)
elbow_k = np.argmin(diffs) + 2  # Simple method to find elbow
plt.axvline(x=elbow_k,color='red',linestyle='--',alpha=0.7,label=f'Suggested k={elbow_k}')
plt.legend()
plt.show()

print(f"Elbow method suggests k={elbow_k} might be optimal for this  data")
print("(Note: Random data often doesn't have clear natural clusters)")

# 5. Analyze Cluster Statistics
print("\n" + "="*50)
print("STEP 5: Cluster Analysis")
print("="*50)

print("Cluster Statistics:")
print("-" * 40)

for i in range(n_clusters):
    cluster_points = X[labels == i]
    centroid = centroids[i]

    print(f"\nCluster {i+1}:")
    print(f"  Number of points: {len(cluster_points)}")
    print(f"  Centroid: ({centroid[0]:.2f}, {centroid[1]:.2f})")
    print(f"  X range: [{cluster_points[:, 0].min():.2f}, {cluster_points[:, 0].max():.2f}]")
    print(f"  Y range: [{cluster_points[:, 1].min():.2f}, {cluster_points[:, 1].max():.2f}]")

    # Calculate distances from centroid
    distances = np.linalg.norm(cluster_points - centroid, axis=1)
    print(f"  Average distance to centroid: {np.mean(distances):.2f}")
    print(f"  Max distance to centroid: {np.max(distances):.2f}")
    print(f"  Min distance to centroid: {np.min(distances):.2f}")

# @title 6. Visualize Clusters Separately
print("\n" + "="*50)
print("STEP 6: Visualizing Individual Clusters")
print("="*50)

# Create subplot for each cluster
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
axes = axes.ravel()

colors = ['red', 'blue', 'green', 'purple']

for i in range(n_clusters):
    cluster_points = X[labels == i]
    axes[i].scatter(cluster_points[:, 0], cluster_points[:, 1],
                   c=colors[i], s=50, alpha=0.7, label=f'Cluster {i+1}')
    axes[i].scatter(centroids[i, 0], centroids[i, 1],
                   marker='X',s=200, c='black',edgecolors='white',linewidths=2,
                   label='Centroid')
    axes[i].set_title(f'Cluster {i+1} ({len(cluster_points)} points)')
    axes[i].set_xlabel('X coordinate')
    axes[i].set_ylabel('Y coordinate')
    axes[i].grid(True, alpha=0.3)
    axes[i].set_xlim(0, 10)
    axes[i].set_ylim(0, 10)
    axes[i].legend()

plt.tight_layout()
plt.show()

# @title 7. Test with Different Number of Clusters
print("\n" + "="*50)
print("STEP 7: Testing with Different k Values")
print("="*50)

# Test different k values
test_k_values = [2, 3, 4, 5]
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
axes = axes.ravel()

for idx, k in enumerate(test_k_values):
    kmeans_test = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels_test = kmeans_test.fit_predict(X)
    centroids_test = kmeans_test.cluster_centers_

    axes[idx].scatter(X[:, 0], X[:, 1],c=labels_test,cmap='viridis',s=50,alpha=0.7)
    axes[idx].scatter(centroids_test[:, 0], centroids_test[:, 1],
                     marker='X', s=200, c='red', edgecolors='black', linewidths=2)
    axes[idx].set_title(f'K-means with k={k} (Inertia: {kmeans_test.inertia_:.2f})')
    axes[idx].set_xlabel('X coordinate')
    axes[idx].set_ylabel('Y coordinate')
    axes[idx].grid(True, alpha=0.3)
    axes[idx].set_xlim(0, 10)
    axes[idx].set_ylim(0, 10)

plt.tight_layout()
plt.show()

# @title 8. Summary
print("\n" + "="*50)
print("SUMMARY")
print("="*50)
print("✓ Generated random 2D points uniformly distributed in [0, 10] range")
print("✓ Applied K-means clustering using Euclidean distance")
print("✓ Visualized clustering results for different k values")
print("✓ Used elbow method to find optimal number of clusters")
print("✓ Analyzed cluster statistics and centroids")
print("\nKey Parameters Used:")
print(f"- Number of clusters (k): {n_clusters}")
print(f"- Number of data points: {n_samples}")
print(f"- Data range: [0, 10] for both coordinates")
print(f"- Distance metric: Euclidean")
print(f"- Final inertia: {kmeans.inertia_:.2f}")
print("\nObservations:")
print("- Random points have no natural clustering structure")
print("- K-means still partitions the space into Voronoi regions")
print("- The elbow method may show a less distinct elbow with random data")
print("\nK-means clustering on random points completed successfully!")
```

```python
# @title K-means Clustering on 2D Points (Using Built-in KMeans)
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

# Set random seed for reproducibility
np.random.seed(42)

# 1. Generate Synthetic 2D Points with make_blobs ==============================
print("="*50)
print("STEP 1: Generating Synthetic 2D Points")
print("="*50)

# Generate the data with 4 natural clusters
n_samples = 300
n_clusters = 4
X, y_true = make_blobs(n_samples=n_samples,
                       centers=n_clusters,
                       cluster_std=1.5,
                       random_state=42)

print(f"Generated {n_samples} points with {n_clusters} natural clusters")
print(f"Data shape: {X.shape}")
print(f"First 5 points:\n{X[:5]}")

# Visualize the generated data
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y_true, cmap='viridis', s=50, alpha=0.7)
plt.title(f'Generated Data with {n_clusters} Natural Clusters', fontsize=14)
plt.xlabel('X coordinate')
plt.ylabel('Y coordinate')
plt.colorbar(label='True Cluster Label')
plt.grid(True, alpha=0.3)
plt.show()

# @title 2. Apply K-means Clustering
print("\n" + "="*50)
print("STEP 2: Applying K-means Clustering")
print("="*50)

# Create and fit KMeans model
kmeans = KMeans(n_clusters=n_clusters,
                random_state=42,
                n_init=10,
                max_iter=300)

# Fit the model to the data
kmeans.fit(X)

# Get the cluster labels and centroids
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

print("K-means Results:")
print(f"Number of iterations: {kmeans.n_iter_}")
print(f"Inertia (sum of squared distances): {kmeans.inertia_:.2f}")
print(f"\nCentroids:")
for i, centroid in enumerate(centroids):
    print(f"Cluster {i+1}: [{centroid[0]:.2f}, {centroid[1]:.2f}]")

# @title 3. Visualize Clustering Results
print("\n" + "="*50)
print("STEP 3: Visualizing Results")
print("="*50)

# Create side-by-side comparison
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Plot 1: Original data with true labels
ax1.scatter(X[:, 0], X[:, 1], c=y_true, cmap='viridis', s=50, alpha=0.7)
ax1.set_title('Original Data (True Labels)', fontsize=12)
ax1.set_xlabel('X coordinate')
ax1.set_ylabel('Y coordinate')
ax1.grid(True, alpha=0.3)

# Plot 2: K-means clustering results
scatter = ax2.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50, alpha=0.7)
ax2.scatter(centroids[:, 0], centroids[:, 1],
           marker='X', s=200, c='red', edgecolors='black', linewidths=2,
           label='Centroids')
ax2.set_title('K-means Clustering Results', fontsize=12)
ax2.set_xlabel('X coordinate')
ax2.set_ylabel('Y coordinate')
ax2.legend()
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# @title 4. Elbow Method to Find Optimal K
print("\n" + "="*50)
print("STEP 4: Elbow Method to Find Optimal Number of Clusters")
print("="*50)

# Calculate inertia for different values of k
k_range = range(1, 11)
inertias = []

for k in k_range:
    kmeans_test = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans_test.fit(X)
    inertias.append(kmeans_test.inertia_)

# Plot elbow curve
plt.figure(figsize=(8, 6))
plt.plot(k_range, inertias, 'bo-', linewidth=2, markersize=8)
plt.xlabel('Number of Clusters (k)', fontsize=12)
plt.ylabel('Inertia (Sum of Squared Distances)', fontsize=12)
plt.title('Elbow Method for Optimal k', fontsize=14)
plt.grid(True, alpha=0.3)
plt.xticks(k_range)

# Highlight the elbow point
plt.axvline(x=n_clusters, color='red', linestyle='--', alpha=0.7, label=f'k={n_clusters}')
plt.legend()

plt.show()

print(f"The elbow point suggests k={n_clusters} is optimal (matching our generated data)")

# @title 5. Analyze Cluster Statistics
print("\n" + "="*50)
print("STEP 5: Cluster Analysis")
print("="*50)

print("Cluster Statistics:")
print("-" * 40)

for i in range(n_clusters):
    cluster_points = X[labels == i]
    centroid = centroids[i]

    print(f"\nCluster {i+1}:")
    print(f"  Number of points: {len(cluster_points)}")
    print(f"  Centroid: ({centroid[0]:.2f}, {centroid[1]:.2f})")

    # Calculate distances from centroid
    distances = np.linalg.norm(cluster_points - centroid, axis=1)
    print(f"  Average distance to centroid: {np.mean(distances):.2f}")
    print(f"  Max distance to centroid: {np.max(distances):.2f}")
    print(f"  Min distance to centroid: {np.min(distances):.2f}")

# @title 6. Test with Different Data (Optional)
print("\n" + "="*50)
print("STEP 6: Testing with Different Data Distribution")
print("="*50)

# Generate data with different characteristics
X2, _ = make_blobs(n_samples=300,
                   centers=3,
                   cluster_std=[1.0, 2.5, 0.5],
                   random_state=42)

# Apply K-means
kmeans2 = KMeans(n_clusters=3, random_state=42, n_init=10)
labels2 = kmeans2.fit_predict(X2)
centroids2 = kmeans2.cluster_centers_

# Visualize
plt.figure(figsize=(8, 6))
plt.scatter(X2[:, 0], X2[:, 1], c=labels2, cmap='plasma', s=50, alpha=0.7)
plt.scatter(centroids2[:, 0], centroids2[:, 1],
           marker='X', s=200, c='red', edgecolors='black', linewidths=2,
           label='Centroids')
plt.title('K-means on Data with Varying Cluster Density', fontsize=14)
plt.xlabel('X coordinate')
plt.ylabel('Y coordinate')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

print(f"Inertia for this dataset: {kmeans2.inertia_:.2f}")

# @title 7. Summary
print("\n" + "="*50)
print("SUMMARY")
print("="*50)
print("✓ Generated 2D synthetic data with natural clusters")
print("✓ Applied K-means clustering using Euclidean distance")
print("✓ Visualized original data vs clustering results")
print("✓ Used elbow method to validate optimal number of clusters")
print("✓ Analyzed cluster statistics and centroids")
print("\nKey Parameters Used:")
print(f"- Number of clusters (k): {n_clusters}")
print(f"- Number of data points: {n_samples}")
print(f"- Distance metric: Euclidean")
print(f"- Final inertia: {kmeans.inertia_:.2f}")
print("\nK-means clustering completed successfully!")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from mpl_toolkits.mplot3d import Axes3D

# Generate 1000 random 3D points
np.random.seed(42)
X = np.random.randn(1000, 3) * 2  # 1000 points in 3D

# Apply K-means with 3 clusters
kmeans = KMeans(n_clusters=3, n_init=10, random_state=42)
labels = kmeans.fit_predict(X)

# Plot in 3D
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot points colored by cluster
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=labels, cmap='viridis', s=50, alpha=0.6)

# Plot centroids
ax.scatter(kmeans.cluster_centers_[:, 0],
           kmeans.cluster_centers_[:, 1],
           kmeans.cluster_centers_[:, 2],
           marker='X', s=200, c='red', label='Centroids')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('K-means Clustering in 3D')
ax.legend()
plt.show()

# Print results
print("Cluster centers (centroids):")
for i, center in enumerate(kmeans.cluster_centers_):
    print(f"Cluster {i+1}: [{center[0]:.2f}, {center[1]:.2f}, {center[2]:.2f}]")
```
