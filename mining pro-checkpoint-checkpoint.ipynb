```python
# Step 2: Rename columns
df = df.rename(columns={
    "OrderID": "Order_ID",
    "CustomerID": "Customer_ID",
    "CustomerAge": "Age",
    "CustomerLifetimeValue": "CLV",
    "DaysSinceRegistration": "Days_Registered",
    "OrderDate": "Date",
    "OrderYear": "Year",
    "OrderMonth": "Month",
    "OrderDayOfWeek": "DayOfWeek",
    "TotalAmount": "Total",
    "FinalAmount": "Final",
    "DiscountAmount": "Discount",
    "DiscountPct": "DiscountPct",
    "PointsEarned": "Points_Earned",
    "PointsRedeemed": "Points_Redeemed",
    "PaymentMethod": "Payment_Method",
    "PaymentStatus": "Payment_Status",
    "ShippingStatus": "Shipping_Status",
    "DeliverySpeed": "Delivery_Speed",
    "ParentCategory": "Parent_Category",
    "CategoryName": "Category",
    "BrandName": "Brand",
    "SizeType": "Size_Type",
    "ColorName": "Color",
    "UnitPrice": "Unit_Price",
    "LineTotal": "Line_Total",
    "TotalSupportTickets": "Support_Tickets",
    "UrgentTickets": "Urgent_Tickets",
    "SLA_BreachedCount": "SLA_Breaches"
})

# Step 3: Fix duplicates or extras
df = df.loc[:, ~df.columns.duplicated()]
for col in ['index','Unnamed: 0']:
    if col in df.columns:
        df = df.drop(columns=[col])

# Step 4: Confirm final count
print(len(df.columns))
print(df.columns.tolist())
```

```python
# Handle Shipping_Status: fill missing with 'Unknown'
df['Shipping_Status'] = df['Shipping_Status'].fillna('Unknown')
df['Shipping_Status'] = df['Shipping_Status'].astype('category')

# Handle Color: fill missing with 'Unknown'
df['Color'] = df['Color'].fillna('Unknown')
df['Color'] = df['Color'].astype('category')

# Check results
print(df[['Shipping_Status','Color']].info())
print(df['Shipping_Status'].value_counts())
print(df['Color'].value_counts().head(10))  # top 10 colors
```

```python
# Convert flag columns (0/1) into boolean True/False
flag_cols = [
    'HasPromotion','IsOnSale','DiscountApplied',
    'HasReview','HasReturn','HasRefund'
]

df[flag_cols] = df[flag_cols].astype(bool)

# Quick check: see how many True values each column has
print(df[flag_cols].sum())   # counts of True
print(df[flag_cols].mean()*100)  # percentage of True
```

```python
import pandas as pd

# --- 1. Identifiers: keep as string --- 
df['Order_ID'] = df['Order_ID'].astype(str)
df['Customer_ID'] = df['Customer_ID'].astype(str)

# --- 2. Dates ---
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# --- 3. Categorical columns ---
categorical_cols = [
    'Gender','AgeGroup','CustomerStatus','DayOfWeek','OrderStatus',
    'City','Country','Payment_Method','Payment_Status','Shipping_Status',
    'Delivery_Speed','Parent_Category','Category','Brand',
    'Size_Type','Color'
]
df[categorical_cols] = df[categorical_cols].astype('category')

# --- 4. Flags (binary indicators) as boolean ---
flag_cols = [
    'HasPromotion','IsOnSale','DiscountApplied',
    'HasReview','HasReturn','HasRefund'
]
df[flag_cols] = df[flag_cols].astype(bool)

# --- 5. Numeric columns ---
numeric_cols = [
    'Age','LoyaltyPoints','CLV','Days_Registered',
    'Year','Month','Total','Final','Discount',
    'DiscountPct','Points_Earned','Points_Redeemed','Quantity','Unit_Price',
    'Line_Total','Support_Tickets','Urgent_Tickets','SLA_Breaches'
]
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors='coerce')

# --- 6. Verify changes ---
print(df.info())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Select all numeric columns
num_cols = df.select_dtypes(include=[np.number]).columns.tolist()
print("Numeric columns:", num_cols)

# Grid layout: 2 plots per column (original + log)
n_cols = 4   # number of plots per row
n_rows = int(np.ceil(len(num_cols) * 2 / n_cols))

plt.figure(figsize=(22, 4 * n_rows))

plot_index = 1
for col in num_cols:
    data = df[col].dropna()
    
    # Original boxplot
    plt.subplot(n_rows, n_cols, plot_index)
    sns.boxplot(x=data, color="skyblue")
    plt.title(f"{col} - Original")
    plt.xlabel("")
    plot_index += 1
    
    # Log-transformed boxplot (only if positive values)
    if (data > 0).all():
        log_data = np.log1p(data)
        plt.subplot(n_rows, n_cols, plot_index)
        sns.boxplot(x=log_data, color="orange")
        plt.title(f"{col} - Log Transformed")
        plt.xlabel("")
        plot_index += 1

plt.tight_layout()
plt.show()
```
