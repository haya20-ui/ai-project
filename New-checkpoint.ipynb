```python
# EXAMPLE: For Loops

# Loop through list
colors = ["red", "green", "blue", "yellow"]
for color in colors:
    print(f"Color: {color}")
```

```python
# Loop with index
for i in range(len(colors)):
    print(f"{i}: {colors[i]}")
```

```python
# Range function
print("Counting to 5:")
for i in range(1, 6):
    print(i, end=" ")
print()
```

```python
# Loop through dictionary
person = {"name": "John", "age": 30, "city": "New York"}
for key in person:
    print(f"{key}: {person[key]}")
```

```python
# Loop through dictionary items
for key, value in person.items():
    print(f"{key} -> {value}")
```

```python
# EXAMPLE: While Loops

# Basic while loop
count = 0
while count < 5:
    print(f"Count: {count}")
    count += 1
```

```python
# While with break and continue
number = 0
while True:
    number += 1
    if number == 3:
        continue  # Skip 3
    if number > 5:
        break  # Stop at 5
    print(number)
```

```python
print(f"3^2 = {power(3)}")      # Uses default
print(f"2^5 = {power(2, 5)}")   # Override default
```

```python
# Function with multiple returns
def calculate_stats(numbers):
    """Calculate mean, min, and max of a list"""
    if not numbers:
        return None, None, None
    
    mean = sum(numbers) / len(numbers)
    minimum = min(numbers)
    maximum = max(numbers)
    return mean, minimum, maximum
```

```python
data = [4, 7, 2, 9, 3]
avg, min_val, max_val = calculate_stats(data)
print(f"Stats: Mean={avg}, Min={min_val}, Max={max_val}")
```

```python
# EXAMPLE: Try-Except Blocks

# Basic error handling
def safe_divide(a, b):
    """Divide two numbers with error handling"""
    try:
        result = a / b
        return result
    except ZeroDivisionError:
        print("Error: Cannot divide by zero!")
        return None
```

```python
print(safe_divide(10, 2))
print(safe_divide(10, 0))
```

```python
# Multiple exception types
def process_input(value):
    """Process user input with multiple error checks"""
    try:
        number = int(value)
        result = 100 / number
        return result
    except ValueError:
        print("Error: Please enter a valid number")
    except ZeroDivisionError:
        print("Error: Cannot divide by zero")
    except Exception as e:
        print(f"Unexpected error: {e}")
    return None
```

```python
# Finally clause
def read_file_safe(filename):
    """Read file with proper error handling"""
    try:
        file = open(filename, 'r')
        content = file.read()
        return content
    except FileNotFoundError:
        print(f"Error: File '{filename}' not found")
        return None
    finally:
        # This always executes
        try:
            file.close()
            print("File closed")
        except:
            pass
```

```python
# EXAMPLE: Reading and Writing Files

# Writing to a file
def write_shopping_list(items, filename="shopping.txt"):
    """Write shopping list to file"""
    try:
        with open(filename, 'w') as file:
            file.write("Shopping List\n")
            file.write("=" * 20 + "\n")
            for i, item in enumerate(items, 1):
                file.write(f"{i}. {item}\n")
        print(f"Shopping list saved to {filename}")
    except Exception as e:
        print(f"Error writing file: {e}")
```

```python
# Reading from a file
def read_shopping_list(filename="shopping.txt"):
    """Read shopping list from file"""
    try:
        with open(filename, 'r') as file:
            content = file.read()
            print("File contents:")
            print(content)
            return content
    except FileNotFoundError:
        print(f"File {filename} not found")
        return None
```

```python
# Test file operations
items = ["Milk", "Bread", "Eggs", "Cheese"]
write_shopping_list(items)
read_shopping_list()
```

```python
# Appending to a file
def add_to_log(message, filename="log.txt"):
    """Append message to log file"""
    from datetime import datetime
    timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    
    with open(filename, 'a') as file:
        file.write(f"[{timestamp}] {message}\n")
    print(f"Message logged to {filename}")
```

```python
def word_counter(filename):
    """
    Count word frequencies in a text file
    Returns dictionary with word counts
    """
    word_freq = {}
    
    try:
        with open(filename, 'r') as file:
            text = file.read().lower()
            
            # Remove punctuation (simple approach)
            for char in '.,;!?":':
                text = text.replace(char, '')
            
            # Split into words
            words = text.split()
            
            # Count frequencies
            for word in words:
                if word in word_freq:
                    word_freq[word] += 1
                else:
                    word_freq[word] = 1
        
        # Sort by frequency
        sorted_words = sorted(word_freq.items(), 
                            key=lambda x: x[1], 
                            reverse=True)
        
        # Display top 10
        print("Top 10 most frequent words:")
        for word, count in sorted_words[:10]:
            print(f"{word}: {count}")
        
        return word_freq
        
    except FileNotFoundError:
        print(f"Error: File '{filename}' not found")
        return {}
```
