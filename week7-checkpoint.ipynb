## 1) Introduction: Why Naive Bayes?

- **Idea:** Model class-conditional feature likelihoods assuming **feature independence** given the class.
- **Variants:**
  - **GaussianNB** — assumes continuous features follow class-conditional Gaussians.
  - **MultinomialNB** — designed for non-negative **count** features (e.g., word counts, pixel intensities).
  - **BernoulliNB** — designed for **binary** features (0/1).
- **Pros:** Fast to train/infer, works well with limited data, strong baselines, interpretable log-probabilities.
- **Cons:** Independence assumption can be unrealistic; needs appropriate feature type/transform per variant.
- **Key knobs:** Additive smoothing `alpha` (for Multinomial/Bernoulli), optional binarization thresholds.


## 2) Setup and Imports

```python
# imports and global settings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import load_digits, load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, Binarizer, StandardScaler
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

import warnings
warnings.filterwarnings('ignore')

RANDOM_STATE = 42
print("Libraries imported. RANDOM_STATE =", RANDOM_STATE)
```


## 3) Dataset A (Instructor): Digits

The **Digits** dataset contains 8×8 grayscale images of handwritten digits with pixel values in `[0, 16]`.  
This makes it suitable for demonstrating **MultinomialNB** (non-negative counts) and **BernoulliNB** (binarized pixels). We also include **GaussianNB** for comparison.


```python
# load Digits and split
digits = load_digits()
X = digits.data  # shape (n_samples, 64), values in [0, 16]
y = digits.target
target_names = [str(i) for i in range(10)]
print("Digits shape:", X.shape, "| Classes:", len(target_names))

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=RANDOM_STATE, stratify=y
)
print("Train/Test:", X_train.shape, X_test.shape)
```

### 3.1 Instructor: Train Gaussian, Multinomial, and Bernoulli NB

```python
# GaussianNB on raw features
gnb = GaussianNB()
gnb.fit(X_train, y_train)
y_hat_g = gnb.predict(X_test)
acc_g = accuracy_score(y_test, y_hat_g)
print("GaussianNB Accuracy:", round(acc_g, 4))

# MultinomialNB expects non-negative features (digits already non-negative).
# Normalize to a common scale (optional) to stabilize counts magnitude.
mms = MinMaxScaler()  # scale to [0,1], then scale up to pseudo-counts
X_train_m = np.round(mms.fit_transform(X_train) * 10.0)
X_test_m  = np.round(mms.transform(X_test) * 10.0)

mnb = MultinomialNB(alpha=1.0)
mnb.fit(X_train_m, y_train)
y_hat_m = mnb.predict(X_test_m)
acc_m = accuracy_score(y_test, y_hat_m)
print("MultinomialNB Accuracy:", round(acc_m, 4))

# BernoulliNB on binarized pixels
binz = Binarizer(threshold=8.0)  # threshold at mid-intensity (0..16)
X_train_b = binz.fit_transform(X_train)  # BernoulliNB can also use binarize=.. param
X_test_b  = binz.transform(X_test)

bnb = BernoulliNB(alpha=1.0)
bnb.fit(X_train_b, y_train)
y_hat_b = bnb.predict(X_test_b)
acc_b = accuracy_score(y_test, y_hat_b)
print("BernoulliNB Accuracy:", round(acc_b, 4))

# Compare
summary = pd.DataFrame([
    ("GaussianNB", acc_g),
    ("MultinomialNB", acc_m),
    ("BernoulliNB", acc_b),
], columns=["Model", "Accuracy"]).sort_values("Accuracy", ascending=False)
print("\nSummary on Digits:")
print(summary.to_string(index=False))

# Plot accuracies
plt.figure(figsize=(5,3.5))
plt.bar(summary["Model"], summary["Accuracy"])
plt.title("Digits — Naive Bayes Variants Accuracy")
plt.xlabel("Model")
plt.ylabel("Accuracy")
plt.tight_layout()
plt.show()
```

### 3.2 Instructor: Smoothing (`alpha`) sweep for Multinomial and Bernoulli

```python
# sweep alpha values
alphas = [0.1, 0.5, 1.0, 2.0, 5.0]
acc_m_list, acc_b_list = [], []

for a in alphas:
    m = MultinomialNB(alpha=a)
    m.fit(X_train_m, y_train)
    acc_m_list.append(accuracy_score(y_test, m.predict(X_test_m)))

    b = BernoulliNB(alpha=a)
    b.fit(X_train_b, y_train)
    acc_b_list.append(accuracy_score(y_test, b.predict(X_test_b)))

alpha_df = pd.DataFrame({
    "alpha": alphas,
    "MultinomialNB": acc_m_list,
    "BernoulliNB": acc_b_list
})
print(alpha_df.to_string(index=False))

plt.figure(figsize=(6,3.5))
plt.plot(alphas, acc_m_list, marker='o', label="MultinomialNB")
plt.plot(alphas, acc_b_list, marker='o', label="BernoulliNB")
plt.title("Digits — Accuracy vs alpha")
plt.xlabel("alpha")
plt.ylabel("Accuracy")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

### 3.3 Instructor: Confusion Matrix for Best Variant

```python
# Pick best variant based on accuracy above (simple comparison)
best_name, best_pred = "GaussianNB", y_hat_g
best_acc = acc_g

if acc_m >= best_acc:
    best_name, best_pred, best_acc = "MultinomialNB", y_hat_m, acc_m
if acc_b >= best_acc:
    best_name, best_pred, best_acc = "BernoulliNB", y_hat_b, acc_b

print("Best variant on Digits:", best_name, "Accuracy:", round(best_acc, 4))

cm = confusion_matrix(y_test, best_pred)
print("Confusion matrix (best):\n", cm)

plt.figure(figsize=(4,4))
plt.imshow(cm, interpolation='nearest')
plt.title(f"Digits — Confusion Matrix ({best_name})")
plt.colorbar()
tick_marks = np.arange(len(target_names))
plt.xticks(tick_marks, target_names, rotation=45)
plt.yticks(tick_marks, target_names)
plt.xlabel("Predicted")
plt.ylabel("True")
plt.tight_layout()
plt.show()
```
