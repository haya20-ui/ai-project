```python
# EXAMPLE: Missing Data Operations

# Create data with missing values
df_missing = pd.DataFrame({
    'A': [1, 2, np.nan, 4, 5],
    'B': [5, np.nan, np.nan, 8, 9],
    'C': [10, 11, 12, 13, np.nan],
    'D': [np.nan, 16, 17, 18, 19]
})
```

```python
print("Data with missing values:")
print(df_missing)
```

```python
# Check for missing values
print("\nMissing values per column:")
print(df_missing.isnull().sum())
```

```python
print("\nPercentage of missing values:")
print((df_missing.isnull().sum() / len(df_missing)) * 100)
```

```python
# Drop missing values
df_dropped = df_missing.dropna()  # Drop rows with any NaN
print("\nAfter dropping NaN rows:")
print(df_dropped)
```

```python
df_dropped_col = df_missing.dropna(axis=1)  # Drop columns with any NaN
print("\nAfter dropping NaN columns:")
print(df_dropped_col)
```

```python
# Fill missing values
df_filled_zero = df_missing.fillna(0)
print("\nFilled with 0:")
print(df_filled_zero)
```

```python
df_filled_mean = df_missing.fillna(df_missing.mean())
print("\nFilled with mean:")
print(df_filled_mean)
```

```python
# Forward fill and backward fill
df_ffill = df_missing.fillna(method='ffill')
print("\nForward fill:")
print(df_ffill)
```

```python
# Creating a sample Titanic-like dataset
np.random.seed(42)

titanic_data = pd.DataFrame({
    'PassengerId': range(1, 101),
    'Survived': np.random.choice([0, 1], 100),
    'Pclass': np.random.choice([1, 2, 3], 100),
    'Name': ['Passenger_' + str(i) for i in range(1, 101)],
    'Sex': np.random.choice(['male', 'female'], 100),
    'Age': np.random.randint(1, 80, 100).astype(float),
    'SibSp': np.random.randint(0, 5, 100),
    'Parch': np.random.randint(0, 5, 100),
    'Fare': np.random.uniform(5, 500, 100)
})
```

```python
# Add some missing values
missing_indices = np.random.choice(100, 20, replace=False)
titanic_data.loc[missing_indices, 'Age'] = np.nan

print("Titanic Dataset Sample:")
print(titanic_data.head())
print(f"\nDataset shape: {titanic_data.shape}")
```

```python
# EXAMPLE: Complete Analysis

# 1. Inspect the data
print("\nDataset Info:")
print(titanic_data.info())
```

```python
print("\nMissing Values:")
print(titanic_data.isnull().sum())
```

```python
print("\nStatistical Summary:")
print(titanic_data.describe())
```

```python
# 2. Clean missing values
print(f"\nMissing ages: {titanic_data['Age'].isnull().sum()}")
```

```python
# Fill missing ages with median
titanic_data['Age_filled'] = titanic_data['Age'].fillna(titanic_data['Age'].median())
```

```python
# 3. Feature engineering
titanic_data['FamilySize'] = titanic_data['SibSp'] + titanic_data['Parch'] + 1
titanic_data['IsAlone'] = (titanic_data['FamilySize'] == 1).astype(int)
```

```python
# 4. Groupby analysis
survival_by_class = titanic_data.groupby('Pclass')['Survived'].mean()
print("\nSurvival rate by class:")
print(survival_by_class)
```

```python
survival_by_sex = titanic_data.groupby('Sex')['Survived'].mean()
print("\nSurvival rate by sex:")
print(survival_by_sex)
```

```python
# 5. Filtering
children = titanic_data[titanic_data['Age_filled'] < 18]
print(f"\nChildren on board: {len(children)}")
print(f"Children survival rate: {children['Survived'].mean():.2%}")
```

# PERFORMANCE TIPS
# ================================================

"""
NumPy Performance Tips:
1. Use vectorized operations instead of loops
2. Pre-allocate arrays when possible
3. Use appropriate data types (int32 vs int64)
4. Avoid copying arrays unnecessarily
5. Use views instead of copies when possible

Pandas Performance Tips:
1. Use vectorized string operations (.str accessor)
2. Use .loc and .iloc for indexing
3. Avoid iterrows() when possible
4. Use categorical data types for repeated strings
5. Chain operations efficiently
6. Use query() for complex filtering
"""

```python
# Example: Vectorization vs Loop
import time

# Slow way (loop)
def slow_operation(arr):
    result = []
    for x in arr:
        result.append(x ** 2 + 2 * x + 1)
    return np.array(result)

# Fast way (vectorized)
def fast_operation(arr):
    return arr ** 2 + 2 * arr + 1

# Test
test_array = np.random.rand(100000)

# Time the operations
start = time.time()
slow_result = slow_operation(test_array)
print(f"Loop time: {time.time() - start:.4f} seconds")

start = time.time()
fast_result = fast_operation(test_array)
print(f"Vectorized time: {time.time() - start:.4f} seconds")
```
